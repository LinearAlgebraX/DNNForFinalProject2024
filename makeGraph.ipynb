```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np
# import matplotlib.font_manager as fm

datasetName = ["MNIST", "GTSRB", "Fashion", "CIFAR10", "CIFAR100"]

Cy = []
Py = []
Cx = []
Px = []

for i in datasetName:
    Cfile_name = f"expResult/{i}/cleanLog.txt"
    Pfile_name = f"expResult/{i}/poisonedLog.txt"

    C = np.loadtxt(Cfile_name)
    P = np.loadtxt(Pfile_name)

    tempC = C[:]
    tempP = P[:]
    Cy.append(tempC)
    Py.append(tempP)

    Cxtemp = list(range(1,len(tempC)+1))
    Pxtemp = list(range(1,len(tempP)+1))
    Cx.append(Cxtemp)
    Px.append(Pxtemp)


plt.figure(figsize=(20, 13))
plt.figure(1)

ax1 = plt.subplot(221)
ax1.set_title("MNIST")
ax1.plot(Cx[0], Cy[0], label="clean")
ax1.plot(Px[0], Py[0], label="poisoned")
plt.xticks(Cx[0])
plt.xlabel("Epoch")
plt.ylabel("Activation index")
plt.legend()

ax2 = plt.subplot(222)
ax2.set_title("GTSRB")
ax2.plot(Cx[1], Cy[1], label="clean")
ax2.plot(Px[1], Py[1], label="poisoned")
plt.xticks(Cx[1])
plt.xlabel("Epoch")
plt.ylabel("Activation index")
plt.legend()

# ax3 = plt.subplot(223)
# ax3.set_title("Fashion")
# ax3.plot(Cx[2], Cy[2], label="clean")
# ax3.plot(Px[2], Py[2], label="poisoned")
# plt.xticks(Cx[2])
# plt.xlabel("Epoch")
# plt.ylabel("Activation index")
# plt.legend()

ax3 = plt.subplot(223)
ax3.set_title("CIFAR10")
ax3.plot(Cx[3], Cy[3], label="clean")
ax3.plot(Px[3], Py[3], label="poisoned")
plt.xticks(Cx[3])
plt.xlabel("Epoch")
plt.ylabel("Activation index")
plt.legend()

ax4 = plt.subplot(224)
ax4.set_title("CIFAR100")
ax4.plot(Cx[4], Cy[4], label="clean")
ax4.plot(Px[4], Py[4], label="poisoned")
plt.xticks(Cx[4])
plt.xlabel("Epoch")
plt.ylabel("Activation index")
plt.legend()

plt.show()
```
